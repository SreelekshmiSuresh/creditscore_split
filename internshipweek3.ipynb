Dropping Amount_Invested_Monthly

```python
data.drop(['Monthly_Inhand_Salary','Amount_invested_monthly'],axis=1,inplace=True)
```

Data Splitting

```python
X=data.drop('Credit_Score',axis=True)#independent
y=data['Credit_Score']#dependent
```

Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.20,stratify=y)
```

Modelling

Random Forest

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,confusion_matrix,classification_report
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_clf=RandomForestClassifier(criterion='gini',min_samples_split=3,min_samples_leaf=5,max_features='sqrt')
rf_clf.fit(X_train,y_train)
rf_pred=rf_clf.predict(X_test)
print('Accuracy of Random Forest Classifier', accuracy_score(y_test,rf_pred))
print('Precision of Random Forest Classifier', precision_score(y_test,rf_pred,average='weighted'))
print('Recall of Random Forest Classifier:',recall_score(y_test,rf_pred,average='weighted'))
print('F1 score of Random Forest Classifier', f1_score(y_test,rf_pred,average='weighted'))
print("Classification Report: \n", classification_report(y_test, rf_pred))
```

From the previous week we found that random forest has highest accuracy as compared to other models which is 0.79

Data split after reduction

```python
x=data.drop('Credit_Score',axis=True)#independent
y=data['Credit_Score']#dependent
```

Train Test split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.20,stratify=y)
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,confusion_matrix,classification_report
from sklearn.ensemble import RandomForestClassifier
```

Randomforest

```python
rf_cf=RandomForestClassifier(criterion='gini',min_samples_split=3,min_samples_leaf=5,max_features='sqrt')
rf_cf.fit(x_train,y_train)
rf_pred=rf_cf.predict(x_test)
print('Accuracy of Random Forest Classifier', accuracy_score(y_test,rf_pred))
print('Precision of Random Forest Classifier', precision_score(y_test,rf_pred,average='weighted'))
print('Recall of Random Forest Classifier:',recall_score(y_test,rf_pred,average='weighted'))
print('F1 score of Random Forest Classifier', f1_score(y_test,rf_pred,average='weighted'))
print("Classification Report: \n", classification_report(y_test, rf_pred))
```

So here we got the accuracy of random forest as 0.80

In cross validation we are getting the score same as 0.80
